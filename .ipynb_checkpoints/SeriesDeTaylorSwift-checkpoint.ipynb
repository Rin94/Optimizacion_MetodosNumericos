#### Importar Librerias

```python
import math
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
from sympy.functions import sin,cos
```

#### Funciones matematicas

```python
def factorial(n):
    if n<=0:
        return 1
    else:
        return n*factorial(n-1)
        
def taylor(func,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (func.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p

```

#### GrÃ¡ficamos resultados

```python

def plot(f,orden):
    x_lims = [-2*np.pi, 2*np.pi]
    #x_lims = [0, 1]
    x1 = np.linspace(x_lims[0],x_lims[1],100)
    y1 = []
    for j in range(1,orden,2):
        print(j)
        func = taylor(f,0,j)
        print('Taylor en n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    plt.plot(x1,np.sin(x1),label='sin of x')
    plt.xlim(x_lims)
    plt.ylim([-5,5])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor Swift approximation series')
    plt.show()
```

```python
plt.style.use("ggplot")
x = sy.Symbol('x')
f = sin(x)
orden=11
plot(f,orden)
```
