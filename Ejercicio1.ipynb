#### Librerías

```python
import math
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
from sympy.functions import sin,cos
```

#### Matemáticas

```python
def derivative(f,a,method='central',h=0.01):
    if method == 'central':
        return (f(a + h) - f(a - h))/(2*h)
    elif method == 'forward':
        return (f(a + h) - f(a))/h
    elif method == 'backward':
        return (f(a) - f(a - h))/h
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'.")
        

def secondDerivative(f,a,method='central',h=0.01):
    if method == 'central':
        return (f(a + h) - (2*f(x))+f(a - h))/(h**2)
    elif method == 'forward':
        return (f(a + 2*h) - 2*f(a+h)+f(a))/(h**2)
    elif method == 'backward':
        return (f(a) - 2*f(a - h)+f(a-2*h))/(h**2)
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'.")
    
    
```

#### Plotting

```python
def plotting(x,dxdy,string="central"):
    plt.plot(x,dxdy)
    plt.plot(x,y)
    plt.title(string)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.show()
    print(" ")
```

#### Declaracion de funcion

```python
#x = sy.Symbol('x')
x = np.linspace(0,1,100)
fun= lambda x:(2*np.sin(5*x)+(3*(x**3))-(2*(x**2))+(3*(x))-5)
```

#### Resultados

```python
dxdy=derivative(fun,x)
```

```python
y = fun(x)
plt.plot(x,dxdy,label="Central Difference y=f'(x)")
plt.plot(x,y,label='y=f(x)');
plt.title("Funcion con derivada centrada")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
